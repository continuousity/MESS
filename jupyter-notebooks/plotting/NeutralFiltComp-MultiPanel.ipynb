```python
%load_ext autoreload
%autoreload 2

%matplotlib inline
import matplotlib.pyplot as plt
from IPython.display import display,SVG
#import MESS
import numpy as np
import pandas as pd
import collections
```

```python
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)

neut_df = pd.read_csv("../neutral_MESS/SIMOUT.txt", sep="\t", header=0)
print(len(neut_df))
comp_df = pd.read_csv("../competition_MESS/SIMOUT.txt", sep="\t", header=0)
print(len(comp_df))
filt_df = pd.read_csv("../filtering_MESS/SIMOUT.txt", sep="\t", header=0)
print(len(filt_df))
print("Mean R", neut_df.R.mean(), comp_df.R.mean())
print("Std R", neut_df.R.std(), comp_df.R.std())
print("Mean/std reg trait neut", neut_df.mn_regional_traits.mean(), neut_df.mn_regional_traits.std())
print("Mean/std reg trait comp", comp_df.mn_regional_traits.mean(), comp_df.mn_regional_traits.std())
print("Mean/std local trait neut", neut_df.mn_local_traits.mean(), neut_df.mn_local_traits.std())
print("Mean/std local trait comp", comp_df.mn_local_traits.mean(), comp_df.mn_local_traits.std())

print(neut_df.columns.values)
neut_means = neut_df.mean(axis=0).to_frame()
comp_means = comp_df.mean(axis=0).to_frame()
filt_means = filt_df.mean(axis=0).to_frame()
display(neut_means.T)
display(comp_means.T)
display(filt_means.T)
```

```python
lfilt = 0.75
dfs = {"neutral":neut_df, "filtering":filt_df, "competition":comp_df}

for model, df in dfs.items():
    dfs[model] = df[df["_lambda"] > lfilt]
    tmp_df = df["shannon"].apply(np.exp)/df["R"]
    dfs[model] = dfs[model].assign(shannon=tmp_df)
    
print(len(neut_df), len(filt_df), len(comp_df))
axs = neut_df.hist(figsize=(50, 30), label="neutral", alpha=0.6, color='blue', bins=30)
ax = [val for sublist in axs for val in sublist]
_ = filt_df.hist(ax = ax[:40], label="filtering", color='orange', alpha=0.6, bins=30)
_ = comp_df.hist(ax = ax[:40], label="competition", color='red', alpha=0.6, bins=30)
plt.legend()
```
