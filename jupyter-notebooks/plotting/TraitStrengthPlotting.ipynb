```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
strength = 10.
filt_opt = -6.59174568179
def filt(strength = strength, victim_trait=-3.32923208, filt_opt=filt_opt):
    return 1 - (np.exp(-((victim_trait - filt_opt) ** 2)/strength))
fig, ax = plt.subplots(figsize=(10, 7))
xs = np.linspace(0.0001, 50, 100)
for trait in [-2.32923208, -3.32923208, -4.32923208, -5.32923208, -6.32923208]:
    ax.plot(xs, [filt(x, trait) for x in xs], label=str(trait-filt_opt))
plt.legend()
plt.savefig("/tmp/trait_strength.jpg")
    #(victim_trait - filt_opt ** 2)/strength
```
